# 서울시 세대 및 거주자 전처리 

```python
df_ppl = pd.read_table("서울시주민통계.txt")
df_ppl.head()
```

```python
df_ppl = df_ppl[["기간", "자치구", "세대", "인구"]].drop([0,1])
```

```python
df_ppl = df_ppl.reset_index().drop("index", axis = 1)
```

```python
df_ppl.head(30)
```

```python
df_ppl.info()
```

```python
df_ppl["기간"].str.split(".")[0][0]
```

```python
df_ppl["연도"] = df_ppl["기간"].apply(lambda x : x.split(".")[0]).astype(int)
```

```python
df_ppl["분기"] = df_ppl["기간"].apply(lambda x : x.split(".")[1])
```

```python
df_ppl["분기"] = df_ppl["분기"].apply(lambda x : x.replace("/4", "")).astype(int)
                                
```

```python
df_ppl = df_ppl[~df_ppl["자치구"].str.contains("합계")]
```

```python
df_ppl = df_ppl.drop("기간", axis = 1).reset_index().drop("index", axis = 1)
df_ppl
```

```python
df_ppl = df_ppl[["연도","분기","자치구","세대","인구"]]
```

```python
df_ppl["세대"] = df_ppl["세대"].str.replace(",","").astype(int)
df_ppl["인구"] = df_ppl["인구"].str.replace(",","").astype(int)
```

```python
df_ppl[['세대전년대비', '인구전년대비']] = df_ppl.groupby('자치구')[['세대', '인구']].shift(1)
```

```python
df_ppl.tail()
```

```python
df_ppl["세대전년대비"] = df_ppl["세대"] - df_ppl["세대전년대비"]
df_ppl["인구전년대비"] = df_ppl["인구"] - df_ppl["인구전년대비"]
```

```python
df_ppl.to_csv("서울시세대인구.csv", index = False)
```
